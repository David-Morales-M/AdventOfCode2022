## **Parte 1**

```python
# Definimos una tupla con la instrucci√≥n y los espacios a recorrer en esa direcci√≥n
# Creamos 2 funciones, la primera para indicar la direcci√≥n a la que se mueve 
# seg√∫n la instrucci√≥n. La segunda sirve para seguir el movimiento de la cola/fin de
# la cuerda
# Finalmente ejecutamos los movimientos seg√∫n cada instrucci√≥n en nuestro archivo
# de input, teniendo en cuenta las 2 variables (head y tail)

with open("Data.txt") as f:
  lines = f.readlines()
  lines = list(map(lambda x: (x.split(" ")[0], int(x.split(" ")[1])), lines))


def directions(d):
    if d == "R":
        return (1, 0)
    elif d == "L":
        return (-1, 0)
    elif d == "U":
        return (0, 1)
    elif d == "D":
        return (0, -1)

def move_tail(head_coords, tail_coords, change):
    
    former_location = (tail_coords[0], tail_coords[1])

    if abs(tail_coords[0] - head_coords[0]) > 1 or abs(tail_coords[1] - head_coords[1]) > 1:
        if change[0] != 0 and change[1] != 0:
            if tail_coords[0] == head_coords[0]:
                tail_coords[1] += change[1]
            elif tail_coords[1] == head_coords[1]:
                tail_coords[0] += change[0]
            else:
                tail_coords[0] += change[0]
                tail_coords[1] += change[1]
        else:

            tail_coords[0] += change[0]
            tail_coords[1] += change[1]

            if change[0] != 0 and tail_coords[1] != head_coords[1]:
                tail_coords[1] = head_coords[1]
            elif change[1] != 0 and tail_coords[0] != head_coords[0]:
                tail_coords[0] = head_coords[0]
    
    return (tail_coords[0] - former_location[0], tail_coords[1] - former_location[1])

coords = set()
rope_coords = [[0, 0] for i in range(2)]
coords.add(tuple(rope_coords[-1]))

for instr in lines:
  dir_ = instr[0]
  for i in range(instr[1]):
    change = directions(dir_)
    rope_coords[0][0] += change[0]
    rope_coords[0][1] += change[1]
            
    for j in range(1, 2):
      change = move_tail(rope_coords[j - 1], rope_coords[j], change)
      if change == (0, 0):
        break
    coords.add(tuple(rope_coords[-1]))

print('Los espacios por los que paso la cola por lo menos una vez fueron: ' + str(len(coords)))
```
